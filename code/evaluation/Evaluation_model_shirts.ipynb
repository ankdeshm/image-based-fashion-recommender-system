```python
import torch
from torchvision import models, transforms
import torch.nn as nn
from annoy import AnnoyIndex
scene_image_folder = '/content/modified_dataset'
scene_images = os.listdir(scene_image_folder)
weights = models.ResNet50_Weights.IMAGENET1K_V1
model = models.resnet50(weights=weights)

model.fc = nn.Identity()
# print(model)

model.eval()

transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor()
    ])

annoy_index = AnnoyIndex(2048, 'angular')

for i in range(len(scene_images)):
  image = Image.open(os.path.join(scene_image_folder, scene_images[i]))
  input_tensor = transform(image).unsqueeze(0)

  if input_tensor.size()[1] == 3:
    output_tensor = model(input_tensor)
    annoy_index.add_item(i, output_tensor[0])

    if i % 100 == 0:
      print(f"Processed {i} images")

annoy_index.build(n_trees= 20,n_jobs = -1)
annoy_index.save('scene_index.ann')
```

```python
import os

path = '/content/validation_final/'
base_image_rec = {}

for subdir in files:
    subdir_path = os.path.join(path, subdir)
    images = os.listdir(subdir_path)
    recommendations = []
    for image in images:
        image_path = os.path.join(subdir_path, image)
        if not image.startswith("masked_"):
            base_image = image_path
        else:
            recommendations.append(image_path)  # Found a recommendation image

    if base_image:  # If a base image was found
        base_image_rec[base_image] = recommendations
```

```python

import torch
import tensorflow as tf
import tensorflow_hub as hub
from PIL import Image, ImageDraw
import numpy as np
from scipy.spatial import distance

%matplotlib inline
```

```python
model_url = "https://tfhub.dev/tensorflow/efficientnet/lite0/feature-vector/2"
layer = hub.KerasLayer(model_url)
model = tf.keras.Sequential([layer])
```

```python
IMAGE_SHAPE = (200, 200)
metric = 'cosine'
def extract(file):
  file = Image.open(file).convert('L').resize(IMAGE_SHAPE)
  file = np.stack((file,)*3, axis=-1)
  file = np.array(file)/255.0

  embedding = model.predict(file[np.newaxis, ...])
  vgg16_feature_np = np.array(embedding)
  flattended_feature = vgg16_feature_np.flatten()

  return flattended_feature
```

```python
cosine_similarities = {}

for base_image, recommendations in base_image_rec.items():
    base_image_features = extract(base_image)
    similarities = []

    for recommendation in recommendations:
        recommendation_features = extract(recommendation)
        similarity = 1 - distance.cosine(base_image_features, recommendation_features)
        similarities.append(similarity)

    cosine_similarities[base_image] = similarities
```

```python
sum_cosine_similarities = {}

for base_image, similarities in cosine_similarities.items():
    sum_similarity = sum(similarities)
    sum_cosine_similarities[base_image] = sum_similarity/len(similarities)
```
